### Exercise 2

```python
def PVG(C=100,n=10,r=0.05,g=0):
    return C/(r-g)*(1-(1+g)**n/(1+r)**n)
PVG()
```

### Exercise 3

```python
# https://en.wikipedia.org/wiki/Mortgage_calculator
def mortgage_cal():
    print("Welcome you to Mortgage Calculator!")
    print("Please input some data!")
    P= int(input("Please input your principal loan in Euro here: "))
    APR = float(input("Please input your interest rate annually in percentage here: "))
    year = int(input("Please input your payment time in year here: "))
    r = APR/12/100
    n = 12 * year
    print("Your monthly payment would be {:.2f} EUR".format(P*r*(1+r)**n/((1+r)**n-1)))
mortgage_cal()
```

### Exercise 4

```python
def dict_sum(d1,d2):
    new_dict = {}
    keys = set(list(d1.keys())+list(d2.keys()))
    for k in keys:
        new_dict[k] = d1.get(k,0) + d2.get(k,0)
    return new_dict               
d1 = {'a': 100 , 'b': 200 , 'c':300 }
d2 = {'a': 300 , 'b': 200 , 'd': 400 }
dict_sum(d1,d2)
```

### Exercise 5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(100)
df = pd.DataFrame()
for path in range(5):
    s = np.random.normal(0,1,365)
    s = np.cumsum(s)
    df= pd.concat([df,pd.DataFrame(s,columns=[path])],axis=1)
plt.plot(df)
plt.show()
```
